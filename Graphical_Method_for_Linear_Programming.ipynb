```python
import numpy as np
import matplotlib.pyplot as plt

# Define problem parameters
resource_availability = [4760, 20]
decision_variables = ['X', 'Y']
objective_function = [-50, -40]  # Coefficients are negated compared to the previous version
resource_constraints = [[400, 240],
                        [1, 1]]

# Define objective function
def objective(x, y):
    return objective_function[0] * x + objective_function[1] * y

# Define resource constraint equations
def constraint1(x):
    return (resource_availability[0] - resource_constraints[0][0] * x) / resource_constraints[0][1]

def constraint2(y):
    return (resource_availability[1] - resource_constraints[1][0] * y) / resource_constraints[1][1]

# Generate x and y values for plotting
x_values = np.linspace(0, 20, 100)
y_values1 = constraint1(x_values)
y_values2 = constraint2(x_values)

# Plot resource constraints
plt.plot(x_values, y_values1, label='400X + 240Y <= 4760')
plt.plot(x_values, y_values2, label='X + Y <= 20')

# Plot feasible region
x = np.linspace(0, 20, 100)
y = np.minimum(constraint1(x), constraint2(x))
plt.fill_between(x, 0, y, where=(y>=0), color='gray', alpha=0.5, label='Feasible Region')

# Plot objective function
plt.plot(8, 12, 'ro')  # Optimal point
plt.annotate('Optimal', (8, 12), xytext=(9, 13),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.xlabel('X')
plt.ylabel('Y')

# Add labels and legend
plt.title('Linear Programming Graphical Solution')
plt.grid()
plt.legend()
plt.xlim(0, 20)
plt.ylim(0, 20)

plt.show()
```
